```
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
  return (np.maximum(x,0))

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)
```

**function f(x)** = xxx + xx - x -1

```
f1 = x*x*x + x*x - x - 1

plt.plot(x,f1)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron1 = Relu(-5x-7.7)**

```
f2 = -5*x - 7.7 

plt.plot(x,f2)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron2 = Relu(-1.2x-1.3)**

```
f3 = -1.2*x - 1.3

plt.plot(x,f3)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron3 = Relu(1.2x+1)**

```
f4 = 1.2*x + 1

plt.plot(x,f4)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron4 = Relu(1.2x -.2)**

```
f5 = 1.2*x - .2

plt.plot(x,f5)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron5 = Relu(2x-1.1)**

```
f6 = 2*x - 1.1

plt.plot(x,f6)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

**neuron6 = Relu(5x-5)**

```
f7 = 5*x -5

plt.plot(x,f7)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

```
x = np.arange(-2,2,.2)


n1 = relu(-5*x -7.7)
n2 = relu(-1.2*x -1.3)
n3 = relu(1.2*x + 1)
n4 = relu(1.2*x -.2)
n5 = relu(2*x -1.1)
n6 = relu(5*x - 5)

z = -n1 -n2 -n3 +n4 +n5 +n6

plt.plot(x,z)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),x, color = 'black')
plt.show()
```
