**Activation function**

```
import math
import matplotlib.pyplot as plot
import numpy as np
```

**Linear** 

```
import math
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-8, 8, 0.2)
def linear(x):
  a = []
  for item in x:
    a.append(item)
    return a
y = linear(x)
plt.plot(x)
plt.grid()
plt.show()
```

***tanh or Hyperbolic tangent activation function ***

```
def tanh(x, derivative = False):
  if (derivative == True):
     return(1 - (x ** 2))
  return np.tanh(x)

y = tanh(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

**sigmoid**

```
def sigmoid(x):
  a = []
  for item in x:
    a.append(1/(1+math.exp(-item)))
  return a

y = sigmoid(x)
plt.plot(x,y)  
plt.grid()
plt.show()
```

**reLu**

```
def relu(x):
  a = []
  for item in x:
    if item > 0:
      a.append(item)
    else:
      a.append(0)
  return a

y = relu(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

**SoftPlus**

```
def SoftPlus(x):
  a = []
  for item in x:
    a.append(math.log10(math.exp(item)+1))
  return a

y = SoftPlus(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

**Leaky Relu**

```
def LeakyRelu(x):
  a = []
  for item in x:
    if item > 0:
      a.append(item)
    else:
      alpha = 0.2
      a.append((item*alpha))
  return a

y = LeakyRelu(x)
plt.plot(x,y)
plt.grid()
plt.show()
```
